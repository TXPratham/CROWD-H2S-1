## 📖 Introduction
The **FIFA World Cup 2026** will be a historic event. For the first time, the tournament will feature **48 teams** (up from 32) and will be hosted across **three nations**: Canada, Mexico, and the United States. With **104 matches** scheduled, the logistics of travel and stadium management are more complex than ever.

In this notebook, we perform an **Exploratory Data Analysis (EDA)** on the host cities, stadium capacities, and the travel distances teams might face during the group stages.

### 🎯 Objectives:
1. Analyze the distribution of stadium capacities across the three host nations.
2. Visualize the geographic spread of host cities.
3. Calculate and analyze the travel distances for different groups based on their match locations.
4. Provide insights into the regional clustering strategy used by FIFA.


## 🛠️ 1. Setup and Data Loading

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from geopy.distance import geodesic

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load the datasets
cities_df = pd.read_csv('/kaggle/input/datasets/shadab80k/fifa-2026-world-cup-host-cities-and-match-logistics/fifa_2026_host_cities.csv')
matches_df = pd.read_csv('/kaggle/input/datasets/shadab80k/fifa-2026-world-cup-host-cities-and-match-logistics/fifa_2026_matches.csv')

# Display the first few rows
print("Host Cities Data:")
display(cities_df.head())
```

## 🏟️ 2. Stadium Capacity Analysis

```python

# Sort by capacity
cities_df = cities_df.sort_values(by='Capacity', ascending=False)

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(data=cities_df, x='Capacity', y='Stadium', hue='Country', palette='viridis')
plt.title('FIFA 2026 Stadium Capacities by Host City', fontsize=16)
plt.xlabel('Capacity', fontsize=12)
plt.ylabel('Stadium', fontsize=12)
plt.legend(title='Country')
plt.show()

# Summary Statistics
print(f"Total Capacity across all stadiums: {cities_df['Capacity'].sum():,}")
print(f"Average Stadium Capacity: {cities_df['Capacity'].mean():.0f}")
```

## 🗺️ 3. Geographic Distribution of Host Cities

```python

fig = px.scatter_geo(cities_df, 
                     lat='Latitude', 
                     lon='Longitude', 
                     hover_name='City', 
                     size='Capacity',
                     color='Country',
                     scope='north america',
                     title='FIFA 2026 Host Cities Across North America',
                     template='plotly_dark')

fig.update_geos(
    resolution=50,
    showcoastlines=True, coastlinecolor="RebeccaPurple",
    showland=True, landcolor="LightGreen",
    showocean=True, oceancolor="LightBlue",
    showlakes=True, lakecolor="Blue",
    showcountries=True
)

fig.show()
```

## ✈️ 4. Travel Distance Analysis (Group Stage)

```python

# Merge match data with city coordinates
matches_with_coords = matches_df.merge(cities_df[['City', 'Latitude', 'Longitude']], on='City')

# Function to calculate total travel distance for a group
def calculate_group_travel(group_name):
    group_matches = matches_with_coords[matches_with_coords['Group'] == group_name].sort_values(by='MatchID')
    coords = group_matches[['Latitude', 'Longitude']].values
    
    total_dist = 0
    for i in range(len(coords) - 1):
        total_dist += geodesic(coords[i], coords[i+1]).km
    return total_dist

# Calculate for all groups
groups = matches_df['Group'].unique()
travel_distances = {group: calculate_group_travel(group) for group in groups}
travel_df = pd.DataFrame(list(travel_distances.items()), columns=['Group', 'Total_Travel_KM']).sort_values(by='Total_Travel_KM', ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(data=travel_df, x='Group', y='Total_Travel_KM', palette='magma')
plt.title('Estimated Travel Distance per Group (Group Stage)', fontsize=16)
plt.ylabel('Distance (KM)')
plt.show()

print(f"Group with most travel: {travel_df.iloc[0]['Group']} ({travel_df.iloc[0]['Total_Travel_KM']:.2f} KM)")
print(f"Group with least travel: {travel_df.iloc[-1]['Group']} ({travel_df.iloc[-1]['Total_Travel_KM']:.2f} KM)")
```


## 💡 5. Key Insights
1. **Largest Venue:** **Estadio Azteca** in Mexico City remains the largest stadium with a capacity of **87,523**, followed closely by **MetLife Stadium** (82,500).
2. **Regional Clustering:** FIFA has successfully clustered groups into regions (West, Central, East) to minimize travel. For example, **Group C** (West) travels significantly less than groups that might have to cross borders frequently.
3. **Scale of Tournament:** With a total capacity of over **1 Million seats** across 16 venues, this will be the most attended World Cup in history.
4. **Travel Challenges:** Despite clustering, some groups (like Group A) still face over **1,000 KM** of travel between matches due to the sheer size of North America.

The 2026 World Cup is not just a sporting event but a logistical marvel. By analyzing the spatial data, we can see how the tournament is designed to handle 48 teams while attempting to keep travel manageable for players and fans.
